```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from src.research import preprocess_data, train_fare_model
```

```python
df = pd.read_csv('../data/raw/yellow_tripdata_2015-01.csv', nrows=100000)

print(df[['trip_distance', 'total_amount']].describe())

plt.figure(figsize=(8, 6))
sns.heatmap(df[['trip_distance', 'fare_amount', 'tip_amount']].corr(), annot=True)
plt.title("Кореляція ознак")
plt.savefig('../reports/figures/correlation_matrix.png')

df_clean = preprocess_data(df)
model, error = train_fare_model(df_clean)

print(f"Mean Squared Error: {error:.2f}")
print(f"Формула моделі: Fare = {model.coef_[0]:.2f} * Distance + {model.intercept_:.2f}")
```
